```python
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 25)
y = np.cos(x)
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
plt.figure()
plt.plot(x,y,':')
plt.show()
```

```python
plt.figure()
plt.plot(x,y,'X')
plt.show()
```

```python
plt.figure()
plt.plot(x,y,'--o')
plt.show()
```

```python
plt.figure()
plt.plot(x,y,'g-o')
plt.show()
```

```python
plt.figure()
plt.plot(x,y)
plt.xlim(-2,10)
plt.ylim(-2,2)
plt.show()
```

```python
plt.figure()
plt.plot(x,y)
plt.xscale('log')
plt.show()
```

```python
plt.figure()
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
z = np.sin(x)

plt.figure()
plt.plot(x,y, label='cos')
plt.plot(x,z, label='sin')
plt.grid(True)
plt.legend(loc=1)
plt.show()
```

```python
plt.figure()
plt.plot(x,y, label='cos')
plt.plot(x,z, label='sin')
plt.grid(True)
plt.legend(loc=1)
plt.title('Trig is hard', fontsize=16)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
