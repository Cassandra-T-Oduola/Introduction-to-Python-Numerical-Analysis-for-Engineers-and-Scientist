```python
mu = 4.0; beta = 0.2;
samples = np.random.gumbel(mu,beta,10000)
```

```python
plt.figure()
plt.hist(samples)
plt.show()
```

```python
plt.figure()
plt.hist(samples,bins=100)
plt.show()
```

```python
plt.figure()
plt.hist(samples, bins=60, normed=True, edgecolor='black')
plt.show()
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-1,3,100)
x,y = np.meshgrid(x,y)
# x = x.flatten()
# y = y.flatten()
f = (1.0-x)**2 + (100.0*((y-(x**2))**2))
```

```python
plt.figure()
plt.plot(x,y,'o')
plt.show()
```

```python
print(x.shape, x.size, x.ndim)
```

```python
x
```

```python
plt.figure()
plt.contour(x,y,f)
plt.show()
```

```python
plt.figure()
plt.contour(x,y,f,100)
plt.show()
```

```python
plt.figure()
plt.contourf(x,y,f,100)
plt.colorbar()
plt.show()
```

```python
levels= [0, 0.5, 1, 5, 10, 500, 1000, 2000]
plt.figure()
plt.contourf(x,y,f,levels)
plt.colorbar()
plt.show()
```
